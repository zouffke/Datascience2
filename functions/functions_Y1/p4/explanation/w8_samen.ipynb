## Forecast Methods

### Naive
De voorspelde waarde bij naive is gelijk aan de vorige waarde, f$_{i}$ = x$_{i}$ - 1

```python
# De naive waarde kan bepaald worden met volgende functie
forecast(opbrengsten, forecastMethod='naive', forecastAmount=4, output='table')
#of in een grafiek
forecast(opbrengsten, forecastMethod='naive', forecastAmount=4, output='graph')
```

### Average
De voorspelde waarde bij average is gelijk aan het gemiddelde van alle voorgaande waarden, f$_{i}$ = $\frac{1}{n}\sum{x_{1}}$

```python
# De average waarde kan worden bepaald met volgende functie
forecast(opbrengsten, forecastMethod='average', forecastAmount=4, output='table')

#of in een grafiek
forecast(opbrengsten, forecastAmount=4, forecastMethod='average', output='graph')
```

### Moving average
De voorspelde waarde bij hetvoortschrijdend gemiddelde is gelijk aan het gemiddelde van m voorgaande waarden.

```python
# Het voortschrijdend gemiddelde kan worden berekend met gebruik van volgende functie
forecast(opbrengsten, forecastMethod='moving_average', forecastAmount=4, output='table')

#of in een grafiek
forecast(opbrengsten, forecastMethod='moving_average', forecastAmount=4, output='graph')
```

### Lineaire combinatie
De voorspelde waarde bij het lineaire gemiddelde is gelijk aan het gewogen gemiddeld van m voorgaande waarden

```python
# De lineaire combinatie kan worden berekend met volgende functie
forecast(opbrengsten, forecastMethod='linear_combination', forecastAmount=4, output='table')

# Of in een grafiek
forecast(opbrengsten, forecastAmount=4, forecastMethod='linear_combination', output='graph')
```

#### Gewichten

## Seasons

Tijdsgebonden data bestaat uit verschillende componenten.
* Trend, lineair (T)
* Seizoensgebonden, patroon dat herhaalt (S)
* niet-regelmatige componenten, ruis (R)
<br>
<img src="https://i.imgur.com/LmpqVWQ.png">

Als de seizoensgrootte gekend is kan je ook de m bepalen.
* Kwartalen, jaarlijks wederkerend patroon -> m = 4
* wekelijks -> m = 7

Als de seizoensgrootte niet gekend is kan dit gevonden worden via de autocorrelatie

#### autocorrelatie
1. Verplaats de data met een aantal tijdseenheden (1, 2, 3, ...)
2. Berekend de correlatie tussen de verplaatste en de oorspronkelijke data
3. Als de correlatie hoog is, waarschijnlijk de data met 1 seizoen verplaatsts

```python
#Bepalen met volgende functie, in een grafiek
autocorrelatie(opbrengsten)
```

```python
#Als nummer
autocorrelatie(opbrengsten, mode='number')
```

### modellen
We kunnen 2 modellen opstellen.
1. additief
    het patroon en de ruis vergroten niet met de trend
    x$_{i}$ = T(i) + S(i) + R(i)
2. multiplicatief
    Het patroon en de trend zijn afhankelijk van de trend
    x$_{i}$ = T(i) * S(i) * R(i)
