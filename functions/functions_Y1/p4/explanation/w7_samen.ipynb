## Correlatie

Een correlatie bevat 2 variabelen. Een onafhankelijke variabele en een afhankelijke.
Data kan worden getoond in een scatterplot

```python
linkedin = pd.read_csv('datasets/linkedIn.csv')
#inlezen van de csv in pandas
linkedin
```

```python
##Met alle verplichte parameters
scatterPlot(linkedin, linkedin.loon, linkedin.connecties)

##Met alle parameters ingevuld
scatterPlot(linkedin, linkedin.loon, linkedin.connecties, title='Coole grafiek', grid='-', marker='d', xDivide=[True, 'b', ':'], yDivide=[True, 'g', 'solid'])
```

### CorrelatiecoÃ«fficient

```python
#Geeft enkel de waardes terug van de correlatiecoÃ«fficiÃ«nt
correlatiecoefficient(linkedin)

#Laat de z scores van de x en y as zien, minimum parameters
correlatiecoefficient(linkedin, mode='graph', onafhankelijk=linkedin.connecties, afhankelijk=linkedin.loon)
#alle parameters
correlatiecoefficient(linkedin, afhankelijk=linkedin.loon, onafhankelijk=linkedin.connecties, divide=False, grid='dotted', mode='graph')

#Toont de correlatieCoÃ«fficiÃ«nt met alle punten ingekleurd. Deze functie kan even duren.
correlatiecoefficient(linkedin, mode='graph', onafhankelijk=linkedin.connecties, afhankelijk=linkedin.loon, colored=True)
```

Het bovenstaanded diagram heeft 4 kwadranten,
1. 2 groene
        De groene punten leveren een positieve bijdrage aan rxy
2. 2 rode
        De rode punten leveren een negatieve bijdrage aan rxy

Een rxy van < 0 --> negatief verband.
    Als de ene waarde groter wordt, wordt de andere waarde kleiner en vica versa.

Een rxy van 0 --> Geen verband aanwezig
    Als de ene waarde groter kan de andere waarde groter of kleiner worden, er is geen verband te vinden.

Een rxy van > 0 --> Positief verband.
    Als de ene waarde groter wordt zal de andere waarde ook groter worden en vica versa.

## Lineaire regressie

```python
#functie om te voorspellen wat de waarde wordt van de afhankelijke variabele aan de hand van de onafhankelijke variabele
linRegressie(linkedin, linkedin.loon, linkedin.connecties, valuesToPredict=[300, 258, 44]) #Array is het aantal connecties in dit voorbeeld, het resultaat is het loon
#of
linRegressie(linkedin, linkedin.loon, linkedin.connecties, mode='prediction', valuesToPredict=[300, 258, 44])

#De functie kan ook gebruikt worden om de functie van de rechte af te drukkken
linRegressie(linkedin, linkedin.loon, linkedin.connecties, mode='function')

#De functie kan ook gebruikt worden om de scatterplot te tonen met daarop de rechte die voorkomt uit de voorspelling.
linRegressie(linkedin, linkedin.loon, linkedin.connecties, mode='graph')
```

```python
#de functie kan ook gebruikt worden om de Standard error te weergeven
linRegressie(linkedin, linkedin.loon, linkedin.connecties, mode='se')

#De standard error kan ook in de scatter plot worden geplot
linRegressie(linkedin, linkedin.loon, linkedin.connecties, mode='graph', interval=True)
```
